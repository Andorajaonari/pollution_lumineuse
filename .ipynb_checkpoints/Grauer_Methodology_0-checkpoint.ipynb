```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import seaborn as sns
import datetime
from scipy.interpolate import UnivariateSpline
from astropy.coordinates import AltAz, EarthLocation, SkyCoord, ICRS
from astropy.time import Time
import astropy.units as u
```

```python
df = pd.read_csv('../ID003.csv')
```

```python
df['UTC_DATE'] = pd.to_datetime(df['UTC_DATE'], errors='coerce')
df['LOCTIME']= pd.to_datetime(df['LOCTIME'], errors='coerce')
df['LOC_DATE'] = df['UTC_DATE'] + pd.Timedelta(hours=4)
df = df.set_index('LOC_DATE') #Mettre l'index a l'heure de la Reunion

df['ra_zenith_h'] = df['ALPHAZEN'] / 15.0 #Conversion de RA en heure
```

```python
df['dt'] = (df['T_AMB'] - df['T_SKY']).copy()

#Definition d'un ciel nocturne, sans lune et claire

seuil_soleil = -18
seuil_illumination = 0.05  # Pourcentage d'illumination max (5%)
seuil_altitude_lune = -10  # Altitude max de la lune (0 pour sous l'horizon)
seuil_dt = 20 # ΔT superieure à 20°

night = df[
    (df['ALT_SUN'] < seuil_soleil) &
#    (df['ILLUMOON'] < seuil_illumination) &
    (df['ALT_MOON'] < seuil_altitude_lune) &
    (df['dt'] > seuil_dt)
].copy()

night.index = pd.to_datetime(night.index)
```

```python
#Distribution de la brillance du ciel sur les 3 ans par rapport à ΔT

plt.figure(figsize=(10, 6))

plt.scatter(night['dt'], night['NSB'], marker='.', s=0.5, color ='red')

plt.suptitle('stars737(Observatoire Atmosphérique du Maïdo)')
plt.title(u'NSB vs ΔT')
plt.xlabel(u'ΔT (deg)')
plt.ylabel("NSB (mag/arcsec²)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
necessary_columns = ['UTC_DATE','UTC_JD','UTC_DATE', 'UTC_JD', 'UTCTIME', 'LOC_JD','LOCTIME', 'T_AMB',
                    'T_SKY', 'FREQ', 'NSB', 'AZ_SUN', 'ALT_SUN', 'AZ_MOON','ALT_MOON', 'AGE_MOON', 'ILLUMOON', 'DISTMOON', 'L_ZEN', 'B_ZEN',
                     'ra_zenith_h', 'BETAZEN','dt']

night_r = night[necessary_columns]
```

```python
#Moyenner les données brutes de 1 minute sur des intervalles de 5 minutes échantillons de 5 minutes.

night_5  = night_r .resample('5min').mean()
```

```python
night_5
```

```python
ra_bins = np.arange(0, 24.5, 0.5)

night_5['ra_bin_h'] = pd.cut(
    night_5['ra_zenith_h'],
    bins=ra_bins,
    labels=ra_bins[:-1], # Utilise le début du bin comme étiquette
    right=False # L'intervalle est [a, b), c'est-à-dire inclus à gauche, exclu à droite
).astype(float)

binned_data = night_5.groupby('ra_bin_h')['NSB'].agg(
    mean_M='mean',    # 'mean' est une chaîne de caractères standard, donc pas besoin de tuple
    all_values=list   # Passez directement le callable 'list' pour obtenir toutes les valeurs du groupe
).reset_index()

quiescent_brightness = []
quiescent_std = []

for index, row in binned_data.iterrows():
    values = np.sort(row['all_values']) # Trier les valeurs de nsb dans le bin
    num_faintest = max(1, int(len(values) * 0.10)) # Au moins 1 point
    faintest_10_percent = values[:num_faintest] # Sélectionner les 10% les plus faibles (magnitudes les plus grandes)

    quiescent_brightness.append(np.mean(faintest_10_percent))
    quiescent_std.append(np.std(faintest_10_percent))

binned_data['quiescent_airglow_brightness'] = quiescent_brightness
binned_data['quiescent_airglow_std'] = quiescent_std


print("\n--- Extrait du DataFrame de données agrégées pour le plot de la Figure 1 ---")
print(binned_data.head())


# --- 3. Créer la Figure 1 ---

plt.figure(figsize=(14, 8))

# 3.1. Tracer les points individuels (moyennes de nsb pour chaque bin)
plt.scatter(
    binned_data['ra_bin_h'],
    binned_data['mean_M'],
    alpha=0.7,          # Transparence pour voir la densité des points
    s=20,               # Taille des points
    color='red',       # Couleur des points
    label='Moyenne NSB par bin de ½h RA'
)

# 3.2. Tracer la courbe de l'airglow au repos
plt.plot(
    binned_data['ra_bin_h'],
    binned_data['quiescent_airglow_brightness'],
    color='green', # Couleur de la courbe
    linewidth=2,
    label='Courbe de l\'Airglow au repos (10% les plus faibles)'
)

# 3.3. Tracer la zone d'erreur (écart-type) autour de la courbe d'airglow au repos
plt.fill_between(
    binned_data['ra_bin_h'],
    binned_data['quiescent_airglow_brightness'] - binned_data['quiescent_airglow_std'],
    binned_data['quiescent_airglow_brightness'] + binned_data['quiescent_airglow_std'],
    color='lightgreen', # Couleur de la zone d'erreur
    alpha=0.4,          # Transparence
    label='$\pm 1 \sigma$ Airglow au repos'
)


# --- 4. Personnaliser le graphique ---

plt.xlabel("Ascension Droite (heures de R.A.)")
plt.ylabel("Brillance du Ciel (mag/arcsec²)")
plt.title("Figure 1: Brillance Instrumentale du Ciel vs. Ascension Droite")
plt.xticks(np.arange(0, 25, 2)) # Afficher les ticks toutes les 2 heures de RA
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.xlim(0, 24) # S'assurer que l'axe X va de 0 à 24
plt.ylim(18, 22) # Ajustez les limites Y selon vos données réelles
plt.gca().invert_yaxis() # Inverser l'axe Y pour que les magnitudes plus faibles (plus brillantes) soient en haut
plt.tight_layout()

plt.gca().invert_xaxis() # Inverser l'axe X pour que la RA aille de 24 à 0

plt.show()
```

```python

# --- Étape 1 : Définir les bins d'Ascension Droite (RA) ---
# Les bins de 0.5 heure de RA, de 0 à 24 (exclusif pour le dernier point)
ra_bins = np.arange(0, 24.5, 0.5)

# --- Étape 2 : Attribuer chaque point de donnée à son bin de RA ---
# Utilisation de pd.cut pour catégoriser les RA
night_5['RA_bin'] = pd.cut(night_5['ra_zenith_h'], bins=ra_bins, right=False, labels=ra_bins[:-1])

# --- Étape 3, 4, 5, 6, 7, 8 : Parcourir chaque bin, extraire les données, calculer le 10% le plus faible, moyenner ---
quiescent_data_points = []
for ra_bin_label in ra_bins[:-1]: # Itérer sur les labels des bins (ex: 0.0, 0.5, ...)
    # Sélectionner les données pour le bin actuel
    subset_for_bin = night_5[night_5['RA_bin'] == ra_bin_label]['NSB']
    
    if not subset_for_bin.empty:
        # Étape 6 : Identifier le seuil des 10% les plus faibles (magnitudes les plus élevées)
        # quantile(0.90) donne la valeur où 90% des données sont en dessous (donc les 10% les plus élevés sont au-dessus)
        # Comme des magnitudes plus grandes signifient plus faibles, cela sélectionne les 10% les plus faibles.
        threshold_90th_percentile = subset_for_bin.quantile(0.90)
        
        # Étape 7 : Sélectionner les données au-delà du seuil (les 10% les plus faibles)
        faintest_10_percent = subset_for_bin[subset_for_bin >= threshold_90th_percentile]
        
        if not faintest_10_percent.empty:
            # Étape 8 : Calculer la moyenne pour le bin
            mean_quiescent_M = faintest_10_percent.mean()
            
            # Étape 9 : Stocker les résultats
            # Le centre du bin est le label + 0.25 (ex: 0.0 + 0.25 = 0.25)
            quiescent_data_points.append({
                'RA_bin_center': ra_bin_label + 0.25,
                'quiescent_M_value': mean_quiescent_M
            })
            
# Convertir la liste de dictionnaires en DataFrame
df_quiescent_points = pd.DataFrame(quiescent_data_points).sort_values('RA_bin_center').reset_index(drop=True)

print("\nPoints d'airglow quiescent calculés (les 'chandelles standard'):")
print(df_quiescent_points.head())
print(f"Nombre de points d'airglow quiescent : {len(df_quiescent_points)}")

# --- Étape 10 : Ajustement de la courbe (réutilisé pour la complétude) ---
#quiescent_curve_func = None
#if len(df_quiescent_points) > 3: # Au moins 4 points pour un polynôme d'ordre 3
    # Utilisation de np.polyfit pour la simplicité
#    coeffs = np.polyfit(df_quiescent_points['RA_bin_center'], df_quiescent_points['quiescent_M_value'], 3)
#    quiescent_curve_func = np.poly1d(coeffs)
#    print(f"\nPolynôme ajusté pour la courbe quiescent: {quiescent_curve_func}")
#else:
#    print("\nPas assez de points pour ajuster une courbe polynomiale d'ordre 3.")

# --- Étape d'ajustement de la courbe avec UnivariateSpline ---
quiescent_curve_func_spline = None

# Assure-toi d'avoir suffisamment de points pour la spline
if len(df_quiescent_points) >= 4: # Spline cubique (k=3) a besoin d'au moins 4 points
    x_data = df_quiescent_points['RA_bin_center'].values
    y_data = df_quiescent_points['quiescent_M_value'].values

    try:
        # Essayer d'abord avec le paramètre de lissage par défaut (s=None)
        # qui tente de trouver un bon lissage basé sur la variance des données.
        spl = UnivariateSpline(x_data, y_data, k=3, s=0.01) # k=3 pour spline cubique
        quiescent_curve_func_spline = spl
        print("\nAjustement spline (k=3, s=None) réussi.")

    except Exception as e:
        print(f"\nErreur lors de l'ajustement de la spline avec s=None : {e}")
        print("Tentative avec un s plus élevé (lissage plus important).")
        # Si l'ajustement par défaut échoue (par exemple, données trop dispersées ou pas assez),
        # tu peux essayer un s plus élevé. La valeur de s dépend de l'échelle de tes données.
        # Une bonne approche est de commencer par s=None et d'ajuster si nécessaire.
        # Pour une brillance en mag/arcsec^2, un petit s est souvent suffisant.
        try:
            # S est la somme des carrés des erreurs résiduelles. Il est lié à la variance de tes points.
            # Une estimation raisonnable de s si tu connais l'erreur typique sur tes points est:
            # s = len(x_data) * (erreur_moyenne_au_carré)
            # Puisque tes points rouges sont des moyennes, leur erreur est plus petite que celle des points bruts.
            # Tu pourrais estimer une erreur typique pour ces points moyens (e.g., 0.01 mag/arcsec^2)^2 * n
            
            # Pour l'exemple, utilisons une valeur arbitraire à ajuster
            s_manual = len(x_data) * (0.01)**2 # Exemple si l'erreur moyenne sur un point est 0.01
            #s_manual = 0.01
            
            spl = UnivariateSpline(x_data, y_data, k=3, s=s_manual)
            quiescent_curve_func_spline = spl
            print(f"Ajustement spline (k=3, s={s_manual:.4f}) réussi après échec du défaut.")
        except Exception as e:
            print(f"Échec de l'ajustement de la spline même avec s manuel : {e}")

else:
    print("\nPas assez de points pour ajuster une spline cubique. Nécessite au moins 4 points.")


# --- Visualisation des résultats (similaire à la Fig. 1) ---
plt.figure(figsize=(14, 7))
plt.scatter(night_5['ra_zenith_h'], night_5['NSB'],
            s=5, alpha=0.3, label='Données M(t) filtrées (chaque point)') # Alpha faible car beaucoup de points

plt.scatter(df_quiescent_points['RA_bin_center'], df_quiescent_points['quiescent_M_value'],
            color='red', marker='o', edgecolor='black', zorder=5, label='Moyenne des 10% les plus faibles par bin RA')

if quiescent_curve_func_spline:
    ra_range_for_plot = np.linspace(0, 24, 500)
    # Pour obtenir les valeurs prédites par la spline, tu appelles l'objet spline comme une fonction
    plt.plot(ra_range_for_plot, quiescent_curve_func_spline(ra_range_for_plot),
             color='blue', linewidth=2.5, linestyle='-', label='Courbe continue de l\'airglow au repos (ajustement Spline)')

plt.xlabel('Ascension Droite au Zénith (Heures)')
plt.ylabel('Brillance du ciel, M(t) (mag/arcsec²)')
plt.title('Calcul de la Courbe de l\'Airglow au repos (Adaptation de Fig. 1 de Grauer et al.)')
plt.gca().invert_yaxis() # Inverser l'axe Y pour que les magnitudes plus élevées (ciel plus sombre) soient en haut
plt.xticks(np.arange(0, 25, 2)) # Étiquettes tous les 2h
plt.xlim(0, 24)
plt.legend(loc='best')
plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()
plt.gca().invert_xaxis() # Inverser l'axe X pour que la RA aille de 24 à 0


plt.show()
```

```python
period['night'] = (period.index - pd.Timedelta(hours=6)).date #Creation d'une colonne night de 18h - 6h
period['night_time'] = ((period.index - pd.to_datetime(period['night'])) / pd.Timedelta(hours=1)) - 18 #Creation de la colonne nuit
#Visualisation de l'evolution du NSB par nuit dans un seul graph

fig, axes = plt.subplots(2, 1, figsize=(8, 8))

for d, g in period.groupby('night'):
    axes[0].errorbar(g['night_time'], g['NSB'], yerr = 0.06, markersize=1 ,elinewidth=1, alpha=0.3, fmt='o', label=d)
    axes[0].scatter(g['night_time'], g['NSB'], marker='o', s=5, label=d)
    axes[1].scatter(g['night_time'], g['T_AMB']-g['T_SKY'], marker='.', s=5, label=d)
    
plt.suptitle('stars737(Observatoire Atmosphérique du Maïdo)')

axes[0].set_title(f"NSB per Night {start} to {end}")
#axes[0].set_xlabel("Hours 18:00 to 6")
axes[0].set_ylabel("NSB (mag/arcsec²)")
axes[0].legend(fontsize='x-small', ncol=3)

axes[1].set_title("Temperature")
axes[1].set_ylabel("$T_{amb} - T_{sky}$ (deg)")
axes[1].legend(fontsize='x-small', ncol=3)

# Format x-axis to show HH:MM
def format_hour(x, _):
    hour = int(x + 18) % 24  # shift back to real time
    minute = int((x % 1) * 60)
    return f"{hour:02d}:{minute:02d}"

plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))  # tick every 2 hours
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(format_hour))

plt.tight_layout()
plt.show()
```
