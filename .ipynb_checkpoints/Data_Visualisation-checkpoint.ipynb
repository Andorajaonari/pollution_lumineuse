```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('../ID003.csv')
```

```python
df['UTC_DATE'] = pd.to_datetime(df['UTC_DATE'], errors='coerce')
df['LOCTIME']= pd.to_datetime(df['LOCTIME'], errors='coerce')
```

```python
df['LOC_DATE'] = df['UTC_DATE'] + pd.Timedelta(hours=4)
```

```python
df = df.set_index('LOC_DATE') #Mettre l'index a l'heure de la Reunion
```

```python
night = df[df['ALT_SUN']<= -18]
night.index = pd.to_datetime(night.index)
```

```python
dates_de_nuit = night.index.date

# Compter le nombre de dates uniques
nombre_de_nuits = pd.Series(dates_de_nuit).nunique()

# Compter le nombre de dates uniques
nombre_de_nuits = pd.Series(dates_de_nuit).nunique()

print(f"Nombre total de nuits (jours avec au moins une observation nocturne) : {nombre_de_nuits}")

# print("\nDates des nuits identifiÃ©es :")
# print(pd.Series(dates_de_nuit).unique())
```

```python
night['dt'] = (night['T_AMB'] - night['T_SKY']).copy()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(night['dt'], night['NSB'], marker='.', s=0.5, color ='red')

plt.suptitle('stars737(Observatoire AtmosphÃ©rique du MaÃ¯do)')
plt.title(u'NSB vs ÎT')
plt.xlabel(u'ÎT (deg)')
plt.ylabel("NSB (mag/arcsecÂ²)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
seuil_illumination = 0.05  # Pourcentage d'illumination max (5%)
seuil_altitude_lune = -10  # Altitude max de la lune (0 pour sous l'horizon)
seuil_betazen = 20
seuil_dt = 20

night_clear = night[
    (night['ILLUMOON'] < seuil_illumination) &
    (night['ALT_MOON'] < seuil_altitude_lune) &
#    (np.abs(night['BETAZEN']) > seuil_betazen) &
    (night['dt'] > seuil_dt)
].copy()
night_clear.index = pd.to_datetime(night_clear.index)
```

```python
dates_de_nuit = night_clear.index.date

# Compter le nombre de dates uniques
nombre_de_nuits = pd.Series(dates_de_nuit).nunique()

# Compter le nombre de dates uniques
nombre_de_nuits = pd.Series(dates_de_nuit).nunique()

print(f"Nombre total de nuits claire (jours avec au moins une observation nocturne) : {nombre_de_nuits}")

print("\nDates des nuits identifiÃ©es :")
print(pd.Series(dates_de_nuit).unique())
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(dt, night_clear['NSB'], marker='.', s=0.5, color ='red')

plt.suptitle('stars737(Observatoire AtmosphÃ©rique du MaÃ¯do)')
plt.title(u'NSB vs ÎT')
plt.xlabel(u'ÎT (deg)')
plt.ylabel("NSB (mag/arcsecÂ²)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
min_nsb_date = night_clear['NSB'].idxmin()
min_nsb = night_clear['NSB'].min()
print(f"La valeur minimal de la brillance nocturne du ciel est Ã©gale Ã  : {min_nsb} et date du {min_nsb_date}")
```

```python
# Enlever les observations du lundi et mardi avant 20h
# Lundi est 0, Mardi est 1
is_monday_or_tuesday = night_clear.index.dayofweek.isin([0, 1])

is_between_18h_and_20h = (night_clear.index.hour >= 18) & (night_clear.index.hour < 21)

condition_to_exclude = is_monday_or_tuesday & is_between_18h_and_20h

night_clear_filtered_time = night_clear[~condition_to_exclude].copy()
```

```python
dt_f = night_clear_filtered_time['T_AMB'] - night_clear_filtered_time['T_SKY']
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(dt_f, night_clear_filtered_time['NSB'], marker='.', s=0.5, color ='red')

plt.suptitle('stars737(Observatoire AtmosphÃ©rique du MaÃ¯do)')
plt.title(u'NSB vs ÎT')
plt.xlabel(u'ÎT (deg)')
plt.ylabel("NSB (mag/arcsecÂ²)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
min_nsb_date = night_clear_filtered_time['NSB'].idxmin()
min_nsb = night_clear_filtered_time['NSB'].min()
print(f"La valeur minimal de la brillance nocturne du ciel est Ã©gale Ã  : {min_nsb} et date du {min_nsb_date}")
```

```python
#date = '2022-06'
start = '2024-09-01 17:00'
end ='2024-11-30 06:00'
#period = night_clear.loc[date] #selection de la periode a visualiser
#period = night_clear.loc[start:end].copy()
period = night_clear_filtered_time.loc[start:end].copy()
```

```python
period['night'] = (period.index - pd.Timedelta(hours=6)).date #Creation d'une colonne night de 18h - 6h
period['night_time'] = ((period.index - pd.to_datetime(period['night'])) / pd.Timedelta(hours=1)) - 18 #Creation de la colonne nuit
#Visualisation de l'evolution du NSB par nuit dans un seul graph

fig, axes = plt.subplots(2, 1, figsize=(14, 10))

for d, g in period.groupby('night'):
    axes[0].errorbar(g['night_time'], g['NSB'], yerr = 0.06, markersize=1 ,elinewidth=0.3, fmt='.', label=d)
    axes[1].scatter(g['night_time'], g['T_AMB']-g['T_SKY'], marker='.', s=5, label=d)
    
plt.suptitle('stars737(Observatoire AtmosphÃ©rique du MaÃ¯do)')

axes[0].set_title(f"NSB per Night {start} to {end}")
#axes[0].set_xlabel("Hours 18:00 to 6")
axes[0].set_ylabel("NSB (mag/arcsecÂ²)")
axes[0].legend(fontsize='x-small', ncol=3)

axes[1].set_title("Temperature")
axes[1].set_ylabel("$T_{amb} - T_{sky}$ (deg)")
axes[1].legend(fontsize='x-small', ncol=3)

# Format x-axis to show HH:MM
def format_hour(x, _):
    hour = int(x + 18) % 24  # shift back to real time
    minute = int((x % 1) * 60)
    return f"{hour:02d}:{minute:02d}"

plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(1))  # tick every 2 hours
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(format_hour))

plt.tight_layout()
plt.show()
```

```python
night_filtered = night_clear_filtered_time

night_filtered['night'] = (night_filtered.index - pd.Timedelta(hours=6)).date #Creation d'une colonne night de 18h - 6h

mask_time_window = (
    (night_filtered.index.hour == 23) & (night_filtered.index.minute == 45) | # Condition pour 23:45
    (night_filtered.index.hour == 0)  & (night_filtered.index.minute == 0)  | # Condition pour 00:00
    (night_filtered.index.hour == 0)  & (night_filtered.index.minute == 15)   # Condition pour 00:15
)

nsb_midnight_window = night_filtered[mask_time_window].copy()

mean_nsb_at_midnight_window = nsb_midnight_window.groupby('night')['NSB'].mean()

# Renomme la sÃ©rie rÃ©sultante pour plus de clartÃ©
mean_nsb_at_midnight_window.name = 'mean_nsb_23h45_00h15_window'

# Convertie l'index 'night' de la Series rÃ©sultante en DatetimeIndex pour faciliter le traÃ§age
mean_nsb_at_midnight_window.index = pd.to_datetime(mean_nsb_at_midnight_window.index)
```

```python
# Trace l'Ã©volution de cette moyenne par nuit
photometer_error = 0.06 # mag/arcsec^2

plt.figure(figsize=(12, 6))

plt.errorbar(
    x=mean_nsb_at_midnight_window.index, 
    y=mean_nsb_at_midnight_window.values, 
    yerr=photometer_error,
    fmt='o',
    label=f'NSB Moyen (23:45 - 00:15) avec erreur $\\pm {photometer_error:.3f}$'
)

plt.suptitle('stars737(Observatoire AtmosphÃ©rique du MaÃ¯do)')

plt.title('Ãvolution du NSB Moyen entre 23:45 et 00:15 par Nuit')
plt.xlabel('Date de la Nuit')
plt.ylabel('NSB Moyen (mag/arcsecÂ²)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```
